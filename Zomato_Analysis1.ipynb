```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.columns
```

```python
country_names=final_df['Country'].value_counts().index
country_values=final_df['Country'].value_counts().values
```

```python
## Pie Chart - top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

Zomato's major bussiness is in India foloowed by US and Uk


```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. rating 4.5 to 4.9 ----> Excellent
2. 4.0 to 4.4 -----> very good
3. 3.5 to 3.9 -----> good
4. 2.5 to 3.4 -----> poor
5. 2.4 to 0.1 -----> poor
6. 0.0 ----> not rated


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

## Observation
1. Not rated count is extremely high
2. Max number of rating is between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
### countries name that given 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```
